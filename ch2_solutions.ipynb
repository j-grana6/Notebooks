## Question 13

Although you can argue for both sides, the more reasonable opinion is to disagree.  A tax on cigarettes will indeed reduce demand (shift down).  Furthermore, it is also reasonable to expect that the price of cigarettes will decrease due to the reduced demand.  However, it is not reasonable to expect that the increase in quantity due to the drop in price will outweigh the drop in quantity due to the tax.  The only way that this would be the case would be if demand were perfectly inelastic (demand curve vertical).

```python
import matplotlib
matplotlib.rcParams['savefig.dpi'] = 2 * matplotlib.rcParams['savefig.dpi']
```

```python
import numpy as np
from matplotlib import pyplot as plt
x= np.arange(0,5,.1)
qd = 5-x
qs = x
qdp = 4-x
plt.plot(x, qs, label='Supply')
plt.plot(x, qd, label = 'Demand')
plt.plot(x, qdp, label ='After tax Demand')
plt.legend(loc=4)
plt.text(1.5, 2.6, 'Q loss due to tax')
plt.arrow(2.5, 2.5, -1, 0, head_width=.1, head_length=.1)
plt.arrow(1.4, 2, .5, 0, head_width = .1, head_length=.1)
plt.text(0, 2.1, 'Q gain from price drop')
plt.show()
```

$Q_d = 300-4P \;$ and $\; Q_s = 3P-120$

$300-4P = 3P-120$

$420 = 7P$

$P = 60 \;$   therefore   $\; Q_s = 3(60)-120=60 \;$  and  $Q\;_d = 300 - 4(60)= 60 \; $ so the before reform equilibrium is $\;P=60\;$ and $\;Q= 60\;$

With the tax reform, the supply function changes to $ Q_s' = 3.2P-120 \;$.  Now we need to get the after reform equilibrium by setting the after reform supply equal to demand:

$3.2P -120 = 300-4P$

$7.2P =420$

$P = 58 \frac{1}{3}\;$ and $Q = 66\frac{2}{3}.\;$  To see how much each consumer saves (per unit) we find the difference in the two equilibrium prices:

Unit Savings = $60 - 58\frac{1}{3} = 1\frac{2}{3}\;$ or $\$1.66$
