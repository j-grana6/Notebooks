```python
x= np.arange(0,5,.1)
qd = 5-x
qs = x
qdp = 4-x
plt.plot(x, qs, label='Supply')
plt.plot(x, qd, label = 'Demand')
plt.ylabel('Price per transaction')
plt.xlabel('Transaction Volume')
plt.plot(x, np.ones(len(x))*1.5)
plt.annotate('(12 Million, $3.00)', xy=(2.5, 2.5), xytext=(3, 4), arrowprops=dict(facecolor='black', shrink=0.05))
plt.text(0, 1.5, 'Ceiling at 1.25')
plt.annotate('5M', xy=(1.5, 1.5), xytext = (1.7, 1.0),arrowprops=dict(facecolor='black', shrink=0.05))
plt.annotate('19M', xy=(3.5, 1.5), xytext = (3.2, 1.0),arrowprops=dict(facecolor='black', shrink=0.05))
plt.text(1.7, .5, 'Shortage of 14 Million')
plt.tick_params(axis= 'both', bottom = 'off', labelbottom='off', labelleft='off')
```

Note that the cost to consumers however is more than just the $\$ 1.25$ transaction cost.  Since there are a shortage of 14 million transactions, the consumer will spend an average 14 minutes looking for an ATM.  At a rate of $\$24$ per hour, 14 minutes is worth $24 \cdot \frac{14}{60}=\$5.60$.  Therefore the total cost to consumers is $ \$1.25 + \$5.60 =\$6.85$.
