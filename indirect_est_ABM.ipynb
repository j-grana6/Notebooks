## The Diffusion Equation

Let $h(t)$ be the probability of adoption at time $t$, let $p$ be the independent probability of adoption, let $q$ be the strength of peer effects and let $A(t)$ be the cumulative proportion of agents who have adopted at time $t$.  Then we have:

$h(t) = p + qA(t)$

# The diffusion process graphically

```python
import numpy as np
from matplotlib import pyplot as plt
import random as random
random.Random(634)

def sim_diffusion(num_agents=10000, p=.03, q=.4, error=0, tlim=10):
    """
    Simulates the diffusion process.  If error is 0,
    the result is deterministic.  If error is type(float) > 0, 
    then error is the standard deviation of the additive error 
    of the probability of adoption.
    """
    agents_left = num_agents
    prop_adopt = 0
    adopters = [0]
    t=0
    while t < tlim:
        p_adopt = p +q*prop_adopt + random.normalvariate(0, error)
        adopt_t = p_adopt*agents_left
        adopters.append(adopt_t)
        agents_left = agents_left - adopt_t
        prop_adopt = 1 - (agents_left / float(num_agents))
        t += 1
    return np.cumsum(adopters)

def plot_diffusion(n, p, q, error=None, tlim=10):
    dp = sim_diffusion(n, p=p, q=q, error=error, tlim=tlim)
    fig, (ax1, ax2, ax3) = plt.subplots(1,3)
    fig.set_size_inches(12,6)
    ax1.plot(dp)
    ax1.set_title('Diffusion Process')
    ax1.set_ylabel('Cumulative Number of Adopters')
    ax1.set_xlabel('Time')
    ax2.plot(np.diff(dp))
    ax2.set_title('Discrete Difference (rate)')
    ax2.set_ylabel('Adopters in time t')
    ax2.set_xlabel('Time')
    ax3.plot(dp[:-1], np.diff(dp))
    ax3.set_title('Empirical Specification')
    ax3.set_ylabel('Adopters in time t')
    ax3.set_xlabel('Cumulative Number of Adopters')
    fig.tight_layout()
    plt.show()
    
plot_diffusion(10000,.03,.4, error = 0, tlim=20)
```

# Empirical Specification and Issues

## The model is cast in continuous time so the solution to the adoption rate is:

 

$\frac{dN(t)}{dt} = pm + (q-p)N(t) - \frac{q}{m}N^2(t)$,

where $N_t$ is the cumulative number of adopters at time $t$ and $m$ is the (finite) population size. 

# The empirical specification becomes


$n_t = \alpha_1 + \alpha_2N_t + \alpha_3N_t^2$

Problem with this specification: **we don't have data on $\frac{dN}{dt}$, we only have the discrete counterpart.**  So how do we recover $p$, $q$ and $m$?

# by estimating the parameters of a binomial distribution.


Recall: $h(t) = p + qA(t)$.  Since $A(t)$ is the cumulative proportion of adopters, we can write $A(t)$ as $\frac{N_t}{m}$

$\implies$ each period contains a sample of draws from a binomial distribution


The ML estimates for $h(t)$ are simply the relative frequencies

   * Estimate $\hat{p}$
   * Estimate $\hat{q}(t)$
   * The estimate for $\hat{q}$ is the weighted average of each estimated $\hat{q}(t)$

# Where are we?

So far, we say that if we know the population size, $m$, we can estimate $p$ and if we know $p$, we can estimate $q$, but what exactly is this finite population size $m$?

# Method of Moments estimation

If we know a population moment that is some function of the parameters, then we can choose the parameters such that the sample moment matches the population moment.
   


   * Suppose we know (assume) the expected value of some function of the underlying population.  In other words, we know $\mathbf{E}(g(X,\beta)) = 0$.
   * (In regression we assume that the $X$'s are uncorrelated with the $\epsilon$'s.  Therefore, $\mathbf{E}[X(y-X\beta)] = 0$)
   * By the law of large numbers (i.e. sample moments converge to their population moments) the sample moment should be "close to" the population moment
   * Choose $\hat{\beta}$ to make the population moment match the sample moment.
   * This estimate is consistient (note the MM is a subset of GMM when size of the parameter vector is equal to the number of moment conditions).

# Method of Simulated Moments

Well... it's the same thing.

* Suppose we observe a population moment (usually a very large sample moments)
* Suppose we know the DGP of the moment up to a set of unknown parameters
* Furthermore, suppose an analytical expression for the sample moment is unavailable but is relatively easy to simulate
* We can estimate the parameter vector by minimizing the distance between the "population moment" and the simulated moment
* Under certain conditions, the estimate is consistent

# A trivial regression example

```python
# OLS via MSM
np.random.seed(634)
from numpy.linalg import pinv

X = sm.add_constant(np.random.standard_normal((100,2))) # Exog
beta = np.array([[1], [2], [3]]) # True params
y = np.dot(X,beta) + np.random.standard_normal(size=(100,1))
OLS_estimates = np.dot(pinv(X), y)
print 'The OLS estimates are \n'
print OLS_estimates, "\n"

def sim_root(estbeta):
    estbeta = estbeta.reshape(3,1) # Input needs to be 1-D
    pred = np.dot(X, estbeta) # "Simulation" Step
    error = y - pred.reshape(-1,1)
    return np.dot(X.T, error).flatten() # match input size

from scipy.optimize import fsolve

MSM_estimates = fsolve(sim_root, np.array([0,0,0]))

print 'The MSM estimates are \n' 
print MSM_estimates.reshape(3,1), "\n"
print "Are they the same? \n" 
print OLS_estimates - MSM_estimates.reshape(3,1) < 10**-10
extended_styles()

```

# MSM in the diffusion model

* The moment to match is the mean adoption time of the data, given by $\frac{1}{N_t}\sum_{t=0}^T(tn_t)$

   * Holding p and q constant, for a value of m **do 1000 times**
       * Simulate the diffusion process for $T=10$
       * Estimate $p$, $q$ and $m$ and store their values
   * Using estimates, compute mean and standard error

```python
from scipy.optimize import fmin
def mc_wrapper(mtrue, etrue = .001, ttrue=10): 
    per_weight = np.arange(1, ttrue+1,1)
    # Generates "observed" data
    obs_data = sim_diffusion(num_agents = mtrue, 
                             error=etrue, tlim=ttrue)
    obs_nt = np.diff(obs_data)
    obs_moment = np.sum(per_weight * obs_nt)/obs_data[-1]
    
    def estimate_params(m):
        """
        This function simulates the diffusion process 
        for any value of m and estimates p and q.
        """
        #  For every m, find \hat{p} and \hat{q}
        phat = obs_data[1] / m
        
        qhat_t = (m / obs_data[1:-1]) * \
            (obs_nt[1:] / (m-obs_data[1:-1]) - phat)
        
        var_q_hat = obs_nt[1:] * \
            (m - obs_data[1:-1]) / (m-obs_data[2:])
        
        denom = np.sum(1/var_q_hat)
        qhat = np.sum(qhat_t * (1/var_q_hat)/denom)
        # Now to simulate with estimated parameters
        sim_data = sim_diffusion(num_agents=m, p=phat, 
                                 q=qhat,error=0, tlim=10)
        
        sim_nt = np.diff(sim_data)
        sim_moment = np.sum(per_weight * sim_nt) / sim_data[-1]
        return (sim_moment - obs_moment)**2

    mhat = fmin(estimate_params, mtrue, 
                maxfun = 10**8, maxiter=10**8, 
                disp=0)[0]
    # Estimated m is the one that minimizes the squared 
    # distance between the observed mean adoption time and 
    # the simulated mean adoption time.
    
    
    # After finding m, back out p and q
    p_opt = obs_data[1]/mhat
    
    q_opt_t = (mhat / obs_data[1:-1])* \
        (obs_nt[1:]/(mhat-obs_data[1:-1]) - p_opt)
    
    var_q_hat_opt = obs_nt[1:] * \
        (mhat - obs_data[1:-1])/(mhat-obs_data[2:])
    
    denom_opt = np.sum(1/var_q_hat_opt)
    q_opt = np.sum(q_opt_t * (1/var_q_hat_opt)/denom_opt)
    return np.array([mhat, p_opt, q_opt])
extended_styles()
```

```python
from multiprocessing import Pool
p=Pool()
ests10000 = np.array(p.map(mc_wrapper, np.ones(1000)*10000))
ests100000 = np.array(p.map(mc_wrapper, np.ones(1000)*100000))
ests1000000 = np.array(p.map(mc_wrapper, np.ones(1000)*1000000))
```
