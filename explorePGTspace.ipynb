# Some Code

```python
import itertools
from adsy.display import extended_styles
strat_1 = list(itertools.product(["A", "B", "C"], ["A", "B", "C"]))
strat_2 = list(itertools.product(["X", "Y", "Z"], ["X", "Y", "Z"], ["X", "Y", "Z"]))
extended_styles()
```

```python
#Utility is a dict keyed by state-player1-player2
utils = {"HAX": (625, 312),
         "HAY": (1000, 200),
         "HAZ": (900, 252),
         "HBX": (550, 687.5),
         "HBY": (300, 350),
         "HBZ": (660, 462),
         "HCX": (427, 762),
         "HCY": (532, 380),
         "HCZ": (504, 504),
         "LAX": (-2000, -1000),
         "LAY": (-1200, -250),
         "LAZ": (-1500, -420),
         "LBX": (-125, -156),
         "LBY": (250, 125),
         "LBZ": (150, 105),
         "LCX": (17, 31), 
         "LCY": (280, 200),
         "LCZ": (210, 210)
}
```

```python
mapdict = {"A": 0, "B":1, "C":2}
def EU(p1strat, p2strat, utils=utils, p1top2map=mapdict):
    highukey = "H"+p1strat[0]+p2strat[mapdict[p1strat[0]]]
    lowukey =  "L"+p1strat[1]+p2strat[mapdict[p1strat[1]]]
    p1eu = .5 * utils[highukey][0] + .5 * utils[lowukey][0]
    p2eu = .5 * utils[highukey][1] + .5 * utils[lowukey][1]
    return p1eu, p2eu
```

```python
stratprofiles = list(itertools.product(strat_1, strat_2))
```

```python
def get_intelligence(profile):
    p1strat = profile[0]
    p2strat = profile[1]
    util = EU(p1strat, p2strat)
    #print util
    p1_numalt = float(len(strat_1))
    p2_numalt = float(len(strat_2))
    num_no_betterp1 = 0
    num_no_betterp2 = 0
    for s in strat_1:
        if EU(s, p2strat)[0] <= util[0]:
            num_no_betterp1 +=1
    for s in strat_2:
        if EU(p1strat, s)[1] <= util[1]:
            num_no_betterp2 +=1
    return (num_no_betterp1-1)/(p1_numalt - 1) * (num_no_betterp2 - 1)/(p2_numalt -1)
        
```

```python
intels = []
for p in stratprofiles:
    intels.append(get_intelligence(p))
```

```python
def get_eq_path(profile):
    p1h = profile[0][0]
    p2h = profile[1][mapdict[p1h]]
    p1l = profile[0][1]
    p2l = profile[1][mapdict[p1l]]
    return p1h+p2h+p1l+p2l
```

```python
observed = [get_eq_path(p) for p in stratprofiles]
```

### Note 1: Strategies that give the same on-equilibrium path observations do not necessarily have the same intelligence

Consider player 1's strategy: 

* $A$ if 1, $B$ if 2 

and the following three profiles by player 2:

* $X$ if $A$, $Y$ if $B$, $X$ if $C$
* $X$ if $A$, $Y$ if $B$, $Y$ if $C$
* $X$ if $A$, $Y$ if $B$, $Z$ if $C$

Of course, these three strategies will give the same observed outcome.  In fact, using the first and the third profile for player 2 are both Nash equilibria.  (They are the "good" equilibria)

Do why doesn't the second strategy of player 2 give an intelligence of 1?  The reason is that if player 2 plays $Y$ when player $1$ plays $C$, then player one will have a higher utility if his strategy has him playing $C$ when the market is state 2.

### Note 2

"Not good" equilibria can be in an equivalence class with other profiles that do not have an intelligence of 1.  For example, consider the equivalence class that has player 1 playing $A$ when nature is in state $1$ and $C$ when nature is in state 2.  Player 2 plays $X$ when he observes $A$ and $Z$ wgen he observes $C$.  Of course there are 3 strategies in this equivalence class.  However, two of these strategies do not have intelligence of 1 because player 2 no longer makes a "threat" and therefore player 1 can earn a higher utility by altering his strategy.

### "Full Game" Intelligence

Here we combine the real intelligence with the full game intelligence described by James as:

1.       Calculate the intelligence of s_L and s_F in the usual way, i.e. on-path, and call it IQ(i,s,on-path). While doing that, keep track of all unvisited information sets for each player, i.e. the set of off-path information sets, H_i.

2.       For each player i, for each unvisited information set h in H_i, and for each state n in h, calculate the intelligence of s_i in information set h given that the state is n, call this IQ(i,s,n).

3.       For each player i, for each unvisited information set h, take the average intelligence across nodes n in h:  IQ(i,s,h)=1/|h| * \sum_{n\in h} IQ(i,s,n)

4.       Take the average of IQ(i,s,h) across unvisited information sets in H_i:  IQ(i,s,off-path) = 1/|H_i| * sum_{h\in H_i} IQ(i,s,h).

5.       Calculate IQ(i,s,full-game) = p*IQ(i,s,on-path) + (1-p)*IQ(i,s,off-path), where p is a parameter we control based on the importance of off-path play. Let’s start with something like p=.5.

### Some more code

```python
def get_off_path_info_sets(profile):
    unvisited = set(["A", "B", "C"]) - set(list(profile[0]))
    return unvisited
```

```python
def get_off_path_intel(profile):
    unvisited = list(get_off_path_info_sets(profile))
    unvisited_intels = []
    for u in unvisited:
        num_no_better_high = 0.
        num_no_better_low = 0.
        p2s_action = profile[1][mapdict[u]]
        utility_high = utils["H" + u + p2s_action][1]
        utility_low = utils["L" + u + p2s_action][1]
        p2s_other_actions = list(set(["X", "Y", "Z"]) - set(p2s_action))
        for other in p2s_other_actions:
            utility_high_other = utils["H" + u + other][1]
            utility_low_other = utils["L" + u + other][1]
            if utility_high_other <= utility_high:
                num_no_better_high +=1
            if utility_low_other <= utility_low:
                num_no_better_low +=1
        unvisited_intels.append(.5 * num_no_better_high/2. + .5* num_no_better_low/2.)

        
    return sum(unvisited_intels)/float(len(unvisited_intels))
```

```python
off_path_intels = [get_off_path_intel(s) for s in stratprofiles]
```

```python
print "Strategy", "                        Intelligence", "              Off Path Intel"
for ix, s in enumerate(stratprofiles):
    print s, '      ', ("%.2f" % intels[ix]), '                    ',  off_path_intels[ix]
```

### Note 1

The "good" equilibria do not have off-path intellifences of 1.  
The reason is that the intellegience is the average of the intelligence that are computed by holding the state constant.  Therefore, if the optimal action in an off path information set is different for at least two of the states, the off path intelligence will be less than 1.
