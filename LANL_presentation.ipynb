# Event-Driven Game Theory for Cyber Attack Detection

## Broad Overview

* If we can predict an attacker's behavior in the most general sense, we can protect networks from not only current attack methods but also attack methods that have not yet been observed. 

* Predicting multiple agent scenarios requires ** Game Theory** 

* Traditional game theory suffers from certain rigidities that make the analysis of large computer networks infeasible

## Stochastic Process Model

* **If** we could specify
    * Message Transmission Rates (background and attacker)
    * Rate of infection upon receiving a malicious message
   
* **Then**
    * We could specify a **generator matrix** that fully describes the state of the system
    * $Q(\lambda) =
\begin{pmatrix} \lambda_{1,1} & \lambda_{1,2} & \cdots & \lambda_{1,n} \\
\lambda_{2,1} & \lambda_{2,2} & \cdots & \lambda_{2,n} \\
\vdots  & \vdots  & \ddots & \vdots  \\
\lambda_{m,1} & \lambda_{m,2} & \cdots & \lambda_{m,n}
\end{pmatrix}$

## The Toy Models

* SFT-Net where each node represents a computer
* Each machine has a set of rates at which it transmits messages in a normal state
* The attacker sends messages at a *pre-specified rate*
* Once a node becomes infected, it stays infected.


## Flow

* Generate $K_1$ realizations of network activity when no attacker is present
* Generate $K_2$ realizations of network activity when an attacker is present
* For each dataset ($K_1 + K_2$), compute the likelihood of the data given no attacker is present (anomaly detector): $L(\theta_C | D)$
* Using $L(\theta_C | D)$ as a classifier, create an **ROC Curve** by plotting true positives on the $x$-axis and false positives on the $y$-axis and varying the threshold.
    * Aside: ROC Curve Crash Course
* For each dataset, compute the likelihood of the data given an attacker is present: $L(\theta_A | D)$
* Using the **Likelihood Ratio** $\frac{L(\theta_C | D)}{L(\theta_A | D)}$ as a classifer and create and **ROC Curve**
