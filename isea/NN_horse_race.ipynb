# Neural Network Horse Race

## Initial Meanderings

The goal of this project is to forecast annual zip code level data.  In fact, there are two problems tangled into one.  The first is to optimally interpolate monthly zip code level data from annual zip code level data and monthly metropolitan level data.  The second task is to forecast zip code level data into the future (and the past).  To better understand the problem, consider one metropolitan statistical association ($MSA$) and the zipcodes ($Z$) that belong to that $MSA$.  The variables $MSA$ and $Z$ represent some variable of interest (i.e. employment) for the metropolitan as a whole and each zipcode, respectively   Initially, we will assume that each $MSA$ is independent so that we only consider one $MSA$ in the initial exploration.

* $MSA$ is a scalar variable and is indexed by time, $t$.  
* $Z$ is a vector variable that contains data for each zip-code.  
* $Z_t$ is generated by an (stochastic) autoregressive process, $Z_t =g(Z_{t-1}, \; Z_{t-2}...)$
* There exists some (*possibly* stochastic) function $h$ such that $MSA_t=h(Z_t)$.  Intuitively, $h$ is the function that aggregates zip-code level data to the metropolitan level.  The stochasticity can be attributed to either imperfect data or simple statistical discrepancy that results from different data collection procedures.  The inverse of $h$ would simply be an de-aggregator function.

### Simple Examples

Suppose the $Z_t$ follows an AR1 process, and each $z_t$ within $Z_t$ are independent (no spatial spillovers),  Then we can write
\begin{equation}
Z_t = \mathbf{\rho}Z_{t-1}' + \mathbf{\epsilon}
\end{equation}

where $Z$ and $\mathbf{\rho}$ are $|Z|\times 1$ vectors and ' denotes vector transpose.  In this simple scenario, we can estimate $\rho_i$ separately for each element of $z_i$ in $Z$ and from this, the optimal forecast is straight forward.  This is **unsatisfactory** for the following reasons:

1) There are likely to be spatial dependencies among zip-codes.

2) If we let the frequency of $t$ be months and we have data for $Z_t$, then we do not have data for $Z_{t-1}$ but only have data for $Z_{t-12}$.

3) There is information in the $MSA_{t-1}$ variable that might help us to predict the value of $Z_t$.

One solution to addressing issue 1 is to change $\mathbf{\rho}$ from a vector to a $|Z| \times |Z|$ matrix.  Then we can do a VAR to get the optimal forecast.  

However, the more relevant and interesting issue is to determine how to incorporate the data from the $MSA$ variable into the optimal forecast.

#### Assuming a simple aggregation function

One way to incorporate information in the MSA variable is to carry out a simple procedure:

1) Assume a functional form for the zipcode data

2) Select and aggregation function and derive its closed form from the zipcode data

3) Forecast  $MSA_{t+k}.$

4) Forecast $Z_{t+k} = h^{-1}(MSA_{t+k})$

This basically turns out to be a two step estimation process.  First, we estimate the parameters of the zipcode level function.  Using this we forecast MSA nd then de-aggregate.

## Working through an example

Suppose: 

1) We have two zip-codes, $z_1$, $z_2$, in one MSA. 

2) The aggregating function is loosely defined by the equations $z_{1,t} + z_{2,t} = MSA_t$  Intuitively, the two zipcodes exactly determine the MSA.  

3) Assume $Z_t$ is a multivariate AR(1) process with 0 correlation.  That is $z_{1,t} = \mu_1 + \rho_1 z_{1, t-1} + \epsilon_1$ and $z_{2,t} = \mu_2 + \rho_2 z_{2, t-1} + \epsilon_2$ where the $\epsilon$ terms are independent.

4) The aggregation function is $z_{1_t} + z_{2,t} = MSA_t$.  That is, there is no statistical discrepency between the zipcode level data and the metro level data.

5) This that the DGP for $MSA_t$ is

\begin{equation}
MSA_t = \mu_2 + \mu_1  + \rho_1 z_{1, t-1}+ \rho_2 z_{2, t-1}   + \epsilon_1 + \epsilon_2
\end{equation}

6) With a bit of hand waving, we can show that the distributions of $z_{1,t}$ and $z_{2,t}$ given $MSA_t$ are normal and derive the mean and the variance.  We can also show that $MSA_t$ follows (at most) an ARMA(2,1) process.

### Forecast procedure:

1) Estimate the process governing $Z_t$ so that we can derive the conditional distirbution of $Z_t$ given $MSA_t$.  

2) Estimate the (ARMA) process governing $MSA_t$.

3) Forecast $MSA_t$ to the period of interest

4) Use the conditional distribution of $Z_t$ given $MSA_t$ amd $Z_{t-1}$ to forecast $Z_t$.

# A simulation Example

```python
import numpy as np
import statsmodels.api as sm
from matplotlib import pyplot as plt
from collections import defaultdict
from matplotlib import rc
## for Palatino and other serif fonts use:
#rc('font',**{'family':'serif','serif':['Palatino']})
rc('text', usetex=True)
%matplotlib inline
from adsy.display import extended_styles
extended_styles()
```

# Results

```python
## Set Parameters
## No intercept Model
mu1=0
mu2=0
st_dev_eps1 = .9
st_dev_eps2 = 1.5
rho1 = .8
rho2 = .8
num_years = 50
num_months = num_years * 12
z10 = 0
z20 = 0
z1_mean = mu1/(1-rho1)
z2_mean = mu2/(1-rho2)

def gen_ar1(x0, rho, mu, st_dev, num_periods):
    ts = [x0]
    for j in xrange(num_periods):
        ts.append(mu + ts[-1]*rho + np.random.normal(scale=st_dev))
    return np.asarray(ts)

def n_period_ahead_arma_2_1_forecast(rho1, rho2, phi1, numperiods, now, yesterday, error, mu):
    fcasted = 0
    while fcasted < numperiods:
        fcast = mu + rho1*now + rho2*yesterday +phi1*error
        yesterday = now
        now = fcast
        error = 0
        fcasted +=1
    return fcast
        
        
num_mc_samps = 500
fcasterror_z1_AR = [] # squared errors
fcasterror_z2_AR = [] # squared errors
# Begin Monte Carlo Simulation
msa_param_col = []
fcasterror_z1_msa = defaultdict(list)
fcasterror_z2_msa = defaultdict(list)
msa_mse = defaultdict(list)
for j in xrange(num_mc_samps):
    #if int(j)/10*10 ==j:
        #print j
    z1 = gen_ar1(z10, rho1, mu1, st_dev_eps1, num_months)
    z2 = gen_ar1(z20, rho2, mu2, st_dev_eps2, num_months)
    msa = z1 + z2
    msa_observed = msa [:-1]
    z1_observed = z1[::12][:-1]
    z2_observed = z2[::12][:-1]
    z1_forecast_true_value = z1[-1]
    z2_forecast_true_value = z2[-1]
    msa_forecast_true_value = msa[-1]
    z1_params = sm.tsa.AR(z1_observed).fit(maxlag=1, trend='nc').params
    z2_params = sm.tsa.AR(z2_observed).fit(maxlag=1, trend='nc').params
    try :
        msa_res = sm.tsa.ARIMA(msa_observed, order=(2,0,1)).fit(trend='nc', disp=0)
        msa_params = msa_res.params
        msa_param_col.append(msa_params)
        z1_predicted =  z1_params[0]*z1_observed[-1]
        z2_predicted = z2_params[0]*z2_observed[-1]
        fcasterror_z1_AR.append((z1_predicted - z1_forecast_true_value)**2)
        fcasterror_z2_AR.append((z2_predicted - z2_forecast_true_value)**2)
        # Now we need to forecast the MSA variable anywhere from 1 to 11 periods 
        # ahead.  We assume the data comes out in march, we need to forecast as if 
        # we were in April, May ... etc
        # This boils down to doing an 11, 10, 9, 8, 7, 6...1 period ahead forecast
        for per_ahead in range(1,12):
            this_period_msa = msa_observed[-per_ahead]
            last_period_msa = msa_observed[-per_ahead - 1]
            this_period_error = msa_res.resid[-per_ahead]
            msa_forecast = n_period_ahead_arma_2_1_forecast(msa_params[0], msa_params[1],
                                                            msa_params[2], per_ahead, 
                                                            this_period_msa, last_period_msa, 
                                                            this_period_error, 0)
            msa_mse[per_ahead].append((msa_forecast - msa_forecast_true_value)**2)
            # We can edit the mean of z1 to be conditional on the current value
            cond_z1_mean = z1_params[0]*z1_observed[-1]
            cond_z2_mean = z2_params[0]*z2_observed[-1]
            z1_forecast = ( cond_z1_mean+ (st_dev_eps1**2/(st_dev_eps1**2 +st_dev_eps2**2))*
                                      (msa_forecast - (cond_z1_mean+cond_z2_mean)))
            z2_forecast = (cond_z2_mean + (st_dev_eps2**2/(st_dev_eps1**2 +st_dev_eps2**2))*
                                      (msa_forecast - (cond_z1_mean+cond_z2_mean)))
            fcasterror_z1_msa[per_ahead].append((z1_forecast - z1_forecast_true_value)**2)
            fcasterror_z2_msa[per_ahead].append((z2_forecast - z2_forecast_true_value)**2)
    except Exception:
        #print 'No convergence'
        pass
```

```python
mse_z1_only = np.mean(fcasterror_z1_AR)
mse_z2_only = np.mean(fcasterror_z2_AR)
z1_mse_msa = [np.mean(val) for key, val in fcasterror_z1_msa.iteritems()]
z2_mse_msa = [np.mean(val) for key, val in fcasterror_z2_msa.iteritems()]
fig, ax = plt.subplots(3,1)
ax[0].scatter(np.arange(1,12,1),z1_mse_msa, label = '$Z_1$ MSE forecast')
ax[0].plot([0, 12], [mse_z1_only, mse_z1_only], label= 'MSE from only considering Z1', color='red')
ax[0].set_xlim(0,12)
ax[1].scatter(np.arange(1,12,1), z2_mse_msa, label='$Z_2$ MSE forecast')
ax[1].plot([0,12], [mse_z2_only,mse_z2_only], label = 'MSE from only considering Z2', color='red')
ax[1].set_xlim(0,12)
ax[2].set_xlabel('Period Ahead Forecast')
ax[0].legend(loc=4)
ax[1].legend(loc=4)
fig.text(1,.5, 'Both processes follow AR(1) with $\mu_1 = \mu_2 = 0$ \n ' + 
         '$\sigma_1 = .9, \sigma_2 =1.5, \\rho_1=.8$ and $\\rho_2 = .8$', size=18)
ax[2].scatter(np.arange(1,12,1), [np.mean(val) for key, val in msa_mse.iteritems()], 
              label='MSE for MSA forecast')
ax[2].legend(loc=4)
#ax[0].set_ylim(.02,.03)
fig.set_figheight(8)
fig.set_figwidth(8)
```
