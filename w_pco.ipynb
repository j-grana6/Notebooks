# Result 1 :  Excess Returns on Settlement Date

For this part, we see if we see any strange behavior on the day of the settlement.  For each stock, we calculate its mean daily returns and the standard deviation of the returns.  We then calculate a standardized return for each stock on the day 
of the settlement.  Therefore, for each stock, we have how many standard deviations away from the average return was the return
on the day of the settlement.  The below histogram shows the distribution of returns on the day of the settlement.

```python
import numpy as np
sd_diff = []
pty = []
for co in permco_nums:
    testcase = data[data.PERMCO ==co]
    meanret = np.mean(testcase['Equal-Weighted Return-incl. dividends'])
    stdret = np.std(testcase['Equal-Weighted Return-incl. dividends'])
    settledates = permcod['Date'][permcod.PERMCO==co].values 
    for dat in settledates:
        retonset = testcase['Equal-Weighted Return-incl. dividends'][testcase.Dates ==dat].values
        for i in range(3):
            if len(retonset) ==0 :
                dat = dat + np.timedelta64(1, 'D')
                retonset = testcase['Equal-Weighted Return-incl. dividends'][testcase.Dates ==dat].values
        sd_diff.extend(list((retonset-meanret)/stdret))
        pty.append(permcod['Total Penalty'][permcod.PERMCO == co].values[0])
```

```python
from matplotlib import pyplot as plt
plt.hist(sd_diff, bins=20)
meanofrets = np.mean(sd_diff)
stdofrets = np.std(sd_diff)
plt.title('Distribution of Returns on Settlement Dats')
```

Now we can test whether the true mean is 0 via a T-test.  The null hypothesis is that the mean standardized excess return on the day of the settlement is equal to 0.  

$H_0: \mu \leq 0$

$H_A: \mu > 0$

The test statistic is $\frac{\bar{x}-\mu_0}{\hat{\sigma}/\sqrt{n}}$

In this case, $\bar{x} = .181, \hat{\sigma} = .902$ and $n=144$  This yields a test statistic of 2.41 and a p-value of .009. Therefore, at the .05 level we can conclude that the mean excess return on the day of settlement is greater than 0. 

# Beta regressions

Remember that $R_r = R_f + \beta(R_f - R_m)$ where $R_r$ is the return on the asset, $R_f$ is the risk free rate and $R_m$ 
is the return on the market.  Of course, $\beta$ is the stocks "beta" and it is sometimes considered a measure of risk or volatility.   Note that if we had data on $R_r$ and $R_m$ and assume that $R_f$ is constant over that period of time, we can form a regression and estimate a stock's beta.  Alas, we do have that information!  

We are not interested simply in estimating a stock's beta, though.  Instead, we want to see if the $\beta$, changes as a result of a settlement.  As a heuristic sketch, a company that has been indicted for ethical issues is likely to lower consumer confidence in that stock.  Therefore, the stock would likely have a higher beta and thuis a higher required return to compensate for the risk.  However, after the justice department settles the issue and the company has cleaned itself up, the market would no longer require such high returns and the beta would decrease. 

To test this hypothesis, we will use the Chow test to determine if there is parameter stability.  In other words, we will run 3 separate regressions of the form

$R_r = \alpha + \beta_1(R_m) $

In the first case, we will run the regression on the entire sample (1 year before the settlement to 1 year after the settlement).  We will then run two more regressions, one for each of the time periods.  The Chow test then compares the residules of what is called a "restricted" and "unrestricted" model to determine if the parameter has changed. 

```python
def compute_chow(res_r, res_u1, res_u2):
    return ((res_r.ssr - (res_u1.ssr +res_u2.ssr))/2)/((res_u1.ssr + res_u2.ssr)/(res_r.nobs-4))
```
